```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x= sp.symbols('x')

f_x= 10 * x - 5
derivative_f_x = sp.diff(f_x, x)
```

```python
derivative_f_x
```

**Спостереження**
Похідна функції f'(x) = 10

```python
def plot_func_and_derivative(x, f_x, d_f_x, x_lim, label_1, label_2):


    # Створення графіку
    plt.figure(figsize=(10, 5))
    plt.plot(x, f_x, label=label_1, color='blue')
    plt.plot(x, d_f_x, label=label_2, color='green')
    
    # Додавання легенди, міток та заголовка
    plt.title("Графік функції та її похідної")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.xlim(x_lim)
    
    plt.show()
```

```python
plot_func_and_derivative(x = np.linspace(0, 10),  f_x= 10 * x - 5,   d_f_x = [10]*50, x_lim = [0, 10], label_1 = "f(x) = 10x - 5", label_2 = "f'(x) = 10")
```

**Спостереження**
Графік фукнції рівномірно зростає, отже, чим більше часу учень витратив на вивчення - тим вища оцінка
